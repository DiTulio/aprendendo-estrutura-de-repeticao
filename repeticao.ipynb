```python
#Ex.4 - Separar valores inseridos pelo usuário em duas listas (par e ímpar)

lista_completa = []
lista_par = []
lista_impar = []

for k in range(10):
    numeros = int(input(f"Digite o {k+1} numero"))

    lista_completa.append(numeros)
    
    if numeros %2 == 0:
        lista_par.append(numeros)
    else:
        lista_impar.append(numeros)

print(lista_completa)
print(lista_par)
print(lista_impar)
```
